```python
import mlxtend # apriori 등을 위함
import numpy as np
import pandas as pd
```

```python
# create data-set
data  = np.array([
    ['milk', 'diaper', 'juice'],
    ['lettuce', 'diaper', 'beer'],
    ['milk', 'lettuce', 'diaper', 'beer'],
    ['lettuce', 'beer']
])
```

```python
# Apriori
```

```python
from mlxtend.preprocessing import TransactionEncoder # 희소행렬을 만들기 위함

te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
te_ary
```

```python
te.columns_
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
%%time
from mlxtend.frequent_patterns import apriori

apriori(df, min_support= 0.5, use_colnames= True)
```

```python
# FP - Growth
```

```python
# 전처리 단계는 apriori와 동일
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df = pd.DataFrame(te_ary, columns= te.columns_)
df
```

```python
%%time
from mlxtend.frequent_patterns import fpgrowth

fpgrowth(df,min_support=0.5,use_colnames=True)
```
